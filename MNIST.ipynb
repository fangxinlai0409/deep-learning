```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import input_data
print("packs loaded")
```

```python
print("Download and Extract MNIST dataset")
mnist=input_data.read_data_sets('data/',one_hot=True)
print
print("type of 'mnist' is %s"%(type(mnist)))
print("number of train data is %d"%(mnist.train.num_examples))
print("number of test data id %d"%(mnist.test.num_examples))
```

```python
print("What does the data of MNIST look loke")
trainimg=mnist.train.images
trainlabel=mnist.train.labels
testimg=mnist.test.images
testlabel=mnist.test.labels
```

```python
nsample=5
randidx=np.random.randint(trainimg.shape[0],size=nsample)
for i in randidx:
    curr_img=np.reshape(trainimg[i,:],(28,28))
    curr_label=np.argmax(trainlabel[i,:])
    plt.matshow(curr_img,cmap=plt.get_cmap('gray'))
    plt.title(""+str(i)+"the Training Data"+"Label is"+ str(curr_label))
    print(""+str(i)+"the Training Data"+"Label is"+str(curr_label))
    plt.show()
```
